````python
# === M3 模組：視覺分析 (Gemini 2.5 最終版) ===
# 因為妳的帳號在經典版 SDK 運作最順暢，我們就繼續用它！
!pip install -q -U google-generativeai

import google.generativeai as genai
from PIL import Image
import json
import os

# 👇 貼上妳那把「新專案」的鑰匙
MY_API_KEY = "AIzaSyB-8Rx2XxqVgYEmZb7wPW2yCJRVTHf7hEk"

def start_m3_final():
    print("🚀 系統啟動中...")

    # 1. 設定連線
    genai.configure(api_key=MY_API_KEY)

    # 根據妳剛剛跑出來的清單，我們選這個最強的！
    target_model = "models/gemini-2.5-flash"

    # 2. 檢查照片
    if not os.path.exists("test.jpg"):
        print("❌ 找不到 'test.jpg'！請確認左側是否有照片並已改名。")
        return

    try:
        img = Image.open("test.jpg")
        print(f"📸 成功讀取照片！準備使用 {target_model} 進行分析...")
    except Exception as e:
        print(f"❌ 照片讀取失敗：{e}")
        return

    # 3. 設定 Prompt (給 AI 的指令)
    prompt = """
    你現在是「律動回聲」系統的視覺分析模組。
    請分析這張照片，提取懷舊遊戲需要的參數。

    請直接回傳純 JSON 格式 (不要 Markdown 標記)，包含：
    - scene: (字串) 場景名稱
    - objects: (陣列) 3個關鍵物件
    - vibe: (字串) 氛圍 (如: 溫馨、熱鬧)
    - year_guess: (字串) 推測年代
    - suggested_music_genre: (字串) 建議搭配的音樂風格
    """

    try:
        # 4. 呼叫 AI (傳送圖片 + 文字)
        model = genai.GenerativeModel(target_model)

        print("🤖 AI 正在看照片中 (約需 3-5 秒)...")
        response = model.generate_content([prompt, img])

        # 5. 整理結果
        print("\n====== 🎉 M3 分析結果 (JSON) ======")

        # 嘗試把結果轉成漂亮的 JSON 顯示
        clean_text = response.text.replace("```json", "").replace("```", "").strip()
        try:
            data = json.loads(clean_text)
            print(json.dumps(data, indent=4, ensure_ascii=False))
        except:
            # 如果轉 JSON 失敗，就直接印出文字
            print(response.text)

        print("========================================")
        print("✅ M3 模組測試成功！這就是我们要給 M4 遊戲組的資料！")
        print("妳可以截這張圖放進專題報告裡了！")

    except Exception as e:
        print(f"❌ 發生錯誤：{e}")
        print("如果是 429 -> 代表妳剛剛測試太頻繁，休息 1 分鐘再按一次就好。")

if __name__ == "__main__":
    start_m3_final()
````

````python
# === 任務二：動態難度調整 (修正模型名稱版) ===

def test_dynamic_difficulty():
    # 記得：不需要重新設定 API Key，直接用上面的就好
    genai.configure(api_key=MY_API_KEY)

    # 🛑 修正點：改用妳帳號裡最強的 2.5 Flash
    model_name = "models/gemini-2.5-flash"
    model = genai.GenerativeModel(model_name)

    # 模擬 M4 傳過來的玩家數據
    player_data = {
        "player_id": "Elder_001",
        "last_game_score": 45,   # 分數很低
        "miss_count": 12,        # 漏接很多
        "reaction_time_avg": "2.5s"
    }

    print(f"📊 接收到玩家數據：{player_data}")
    print(f"🤖 使用 {model_name} 計算最佳難度...")

    prompt = f"""
    你是一個針對長者的復健遊戲設計師。
    請根據這位玩家上一場的表現數據：{player_data}
    決定下一關的難度參數。

    規則：
    1. 如果玩家表現差，降低速度，減少踩點密度。
    2. 如果玩家表現好，稍微增加挑戰。

    請回傳 JSON：
    - next_speed (float): 0.5 (慢) ~ 2.0 (快)
    - spawn_density (int): 同時出現的踩點數量 (1~3)
    - encouragement (string): 給長者的一句鼓勵語
    """

    try:
        response = model.generate_content(prompt)
        print("\n====== ⚖️ 動態難度調整 (JSON) ======")
        print(response.text)
        print("==================================")
        print("✅ 成功！AI 已經變成遊戲企劃了！")

    except Exception as e:
        print(f"❌ 錯誤：{e}")

if __name__ == "__main__":
    test_dynamic_difficulty()
````

````python
# === 任務一：語音轉文字 + 意圖分析 (Gemini 全包版) ===
!pip install -q -U google-generativeai

import google.generativeai as genai
import json
import os

# 👇 貼上妳那把「新專案」的鑰匙
MY_API_KEY = "AIzaSyB-8Rx2XxqVgYEmZb7wPW2yCJRVTHf7hEk"

# === 任務一：語音轉文字 + 意圖分析 (Gemini 2.5 修正版) ===
# 不需要重新 pip install，直接跑下面的就好

def test_audio_analysis():
    print("🎤 啟動語音分析模組 (使用 Gemini 2.5)...")

    genai.configure(api_key=MY_API_KEY)

    # 確保妳有上傳錄音檔，並且名字叫做 test_audio.mp3
    audio_file_path = "test_audio.mp3"

    if not os.path.exists(audio_file_path):
        print(f"❌ 找不到 {audio_file_path}，請確認左側檔案區有沒有這個檔案？")
        return

    print("📤 正在上傳音訊到 Google...")

    try:
        # 上傳檔案
        myfile = genai.upload_file(audio_file_path)
        print(f"✅ 上傳成功！準備開始分析...")

        # 🛑 修正點：這裡改成妳帳號能用的 2.5 Flash
        model = genai.GenerativeModel("models/gemini-2.5-flash")

        prompt = """
        請聽這段語音，並執行以下任務：
        1. 【語音轉文字】：精準打出逐字稿。
        2. 【情感分析】：分析說話者的情緒 (Happy/Sad/Nostalgic/Calm)。
        3. 【JSON封裝】：將結果轉為 JSON 格式。

        回傳格式範例：
        {
            "stt_content": "我小時候喜歡去...",
            "emotion": "Nostalgic",
            "keyword_tags": ["童年", "柑仔店"]
        }
        """

        print("🤖 AI 正在聆聽並分析 (約需 5-10 秒)...")
        response = model.generate_content([prompt, myfile])

        print("\n====== 🗣️ 語音分析結果 (JSON) ======")
        print(response.text)
        print("==================================")

        # 為了節省空間，分析完後可以刪除雲端的暫存檔
        myfile.delete()
        print("✅ 測試完成！")

    except Exception as e:
        print(f"❌ 發生錯誤：{e}")

if __name__ == "__main__":
    test_audio_analysis()
````

````python
# === 任務三：人物去背技術展示 (修正安裝版) ===
# 🛑 關鍵修正：我們改裝 "rembg[cpu]"，這樣才會包含運算引擎 onnxruntime
!pip install "rembg[cpu]"

from rembg import remove
from PIL import Image
import io
import os
import json

def test_background_removal():
    print("🎨 啟動影像處理模組...")

    input_path = "test.jpg" # 確保妳的 test.jpg 還在喔！
    output_path = "test_no_bg.png" # 這是去背後要存的檔名

    # 檢查照片在不在
    if not os.path.exists(input_path):
        print(f"❌ 找不到 {input_path}！請確認左側資料夾有沒有這張照片？")
        return

    try:
        input_image = Image.open(input_path)
        print("📸 成功讀取照片！")

        print("⏳ 正在進行 AI 去背 (第一次執行會下載模型，請稍等 10-20 秒)...")

        # === 魔法發生的地方 ===
        output_image = remove(input_image)

        # 存檔
        output_image.save(output_path)

        print(f"✅ 去背大成功！已存為：{output_path}")
        print("👉 請看左側資料夾，點兩下 'test_no_bg.png' 就能看到去背成果了！")

        # 模擬回傳給 Unity 的資料
        result_json = {
            "original_image": "test.jpg",
            "processed_image": "test_no_bg.png",
            "unity_action": "Use processed_image as Sprite, overlay on Generated Background"
        }
        print("\n====== 🖼️ 影像處理回傳 (JSON) ======")
        print(json.dumps(result_json, indent=4))

    except Exception as e:
        print(f"❌ 失敗：{e}")
        print("💡 如果還是報錯，請點選上方選單的【執行階段】->【重新啟動工作階段】，然後再跑一次。")

if __name__ == "__main__":
    test_background_removal()
````
